# Random Forest

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time

# Flatten the class labels
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

# Initialize the RandomForestClassifier without Grid Search CV
fit_rf = RandomForestClassifier(random_state=40, n_estimators=100, max_depth=None, min_samples_split=2, 
                                min_samples_leaf=1, max_features='auto', bootstrap=True)
np.random.seed(40)

# Fit the RandomForestClassifier with the training data
start = time.time()
fit_rf.fit(x_train, y_train)
end = time.time()
print('Time taken to train the model without Grid Search CV: %0.2f seconds' % (end - start))


```

```python
# Predict on the test set
predictions_rf = fit_rf.predict(x_test)
```

```python
# Calculate and print the accuracy of the model
accuracy_rf_without_cv = accuracy_score(y_test, predictions_rf)
print("Accuracy of model without Grid Search CV: %0.5f" % accuracy_rf_without_cv)
```

```python
# Define parameter grid for GridSearchCV
parameter_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Initialize RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
cv = GridSearchCV(rf_model, param_grid=parameter_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit GridSearchCV
start = time.time()
cv.fit(x_train, y_train)
end = time.time()
print('Time taken to train the model with Grid Search CV: %0.2f seconds' % (end - start))

```

```python
# Get best parameters from GridSearchCV
best_params = cv.best_params_
print('Best Parameters using Grid Search \n', best_params)
```

```python
# Predict using the best estimator
y_pred = cv.predict(x_test)
```

```python
# Calculate accuracy score
accuracy_rf_with_cv = accuracy_score(y_test, y_pred)
print("Accuracy of model with Grid Search CV: %0.5f" % accuracy_rf_with_cv)
```

```python
cv.predict([[329,2.09,1,45.00,12,13.200,70.200]])
```

```python
# Bar chart to compare accuracies
import matplotlib.pyplot as plt
labels = ['Without Grid Search CV', 'With Grid Search CV']
accuracies = [accuracy_rf_without_cv, accuracy_rf_with_cv]
plt.bar(labels, accuracies, color=['blue', 'green'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracies')
plt.ylim(0, 1)
plt.show()
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
knn = KNeighborsClassifier(n_neighbors=5)  # you can choose any number of neighbors
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
knn.predict([[329,2.09,1,45.00,12,13.200,70.200]])
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors': np.arange(1, 31),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_
print("Best parameters found: ", best_params)
y_pred = grid_search.best_estimator_.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracy scores
models = ['Logistic Regression', 'Decision Tree', 'Random Forest','KNN','Naive Bayes','SVC']
without_gscv = [0.97, 0.99, 0.99,0.96,0.96,0.98]
with_gscv = [0.97, 0.99, 0.99,0.97,0.95,0.97]

# Setting the positions and width for the bars
bar_width = 0.35
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(r1, without_gscv, color='b', width=bar_width, edgecolor='grey', label='Without GridSearchCV')
plt.bar(r2, with_gscv, color='g', width=bar_width, edgecolor='grey', label='With GridSearchCV')

# Adding labels and title
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Accuracy', fontweight='bold')
plt.title('Model Accuracy Comparison with and without GridSearchCV', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(models))], models)
plt.ylim(0.9, 1.01)

# Adding the legend
plt.legend()

# Adding accuracy labels on top of the bars
for i in range(len(r1)):
    plt.text(r1[i], without_gscv[i] + 0.005, f"{without_gscv[i]:.2f}", ha='center', fontweight='bold')
    plt.text(r2[i], with_gscv[i] + 0.005, f"{with_gscv[i]:.2f}", ha='center', fontweight='bold')

# Displaying the plot
plt.show()
```
